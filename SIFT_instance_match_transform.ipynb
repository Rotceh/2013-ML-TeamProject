# Real SIFT Feature Transform

## Terminology

- `zwd`: **zodiac word**, total 12 words
- `inst`: **instance**, one data entry. In CSV, it represents a row of data
- `feature`: contain both `kp` and `desc`, output of SIFT. one instance can have multiple features generated.
- `kp`: **keypoint**, coming along with `desc`, the location, scale, angle of the feature.
- `desc`: **description**, one can think of it as *visual word* or a patch of the instance, representing how the feature exactly looks like.
    
Many code are tested/played in `play_SIFT_transform.ipynb` notebook.

## Stage 1

- convert all instances into list of features (kp, desc)
- flatten the list in to one-dim array
- rematch each desc of each instance to the flatten list

## Stage 2

- Restore back the information, each instance will count how many proportion of instances are matched for each group of zwd

Here we use the better dataset `60x60addmean` for SIFT detection

### SIFT on all instance

````python
def grey_convert(wd_array):
    """convert [0, 1] greyscale to [0, 255] (dark, white)"""
    return np.round((1 - wd_array) * 255).astype(np.uint8)
````

Parameter for SIFT feature finding

````python
sigma = 1.4 # default: 1.6
contrastThreshold = 0.03 # default: 0.04
nOctaveLayers = 5  # default: 3
edgeThreshold = 7 # default: 10
sift = cv2.SIFT(
    sigma=sigma, 
    contrastThreshold=contrastThreshold,
    nOctaveLayers=nOctaveLayers,
    edgeThreshold=edgeThreshold
)
````

````python
%%time
kp_bucket = []
desc_bucket = []
for inst in flatten_inst_arary:
    inst.kp_list, inst.desc_list = sift.detectAndCompute(
        grey_convert(train_X[inst.ix].reshape(PIXEL_X, PIXEL_Y)),
        None
    )
    kp_bucket.append(inst.kp_list)
    desc_bucket.append(inst.desc_list)
````

## Stage2 - Matching
Here we choose the following features as output:

- top K %features matched of all instances for each `zwd`
- mean %features matched of all instances for each `zwd`

So we will get `12 * (K + 1)` features in our output data frame.

### Naming

```
z[zwd]_top[k]  :   z1_top1, z1_top2, z1_top3, ...
z[zwd]_mean    :   z1_mean, z2_mean, ...
```

````python
TOP_K = 10
N_ZODIAC = 12
out_dict = {'y': train_Y}
# Top K columns
top_column_names = [
    "z{}_top{}".format(zwd+1, k+1)
    for zwd in range(N_ZODIAC) for k in range(TOP_K)
]
# Mean columns
mean_column_names = [
    "z{}_mean".format(zwd+1) for zwd in range(N_ZODIAC)
]
out_dict.update({
    col_name : 0 for col_name in top_column_names + mean_column_names
})
df_sift_train_out = pd.DataFrame(out_dict)
````

````python
# sort column name for better human readability
df_sift_train_out.columns = (
    ['y'] +
    top_column_names +
    mean_column_names
)   
````

````python
df_sift_train_out.columns
````

FLANN Matcher

````python
# FLANN parameters
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=100)   # or pass empty dictionary

flann = cv2.FlannBasedMatcher(index_params, search_params)
````

````python
%%time
DISTANCE_TH = 250

inst = flatten_inst_arary[0]

# the zwd to deal with
# zwd_inst_list = zwd_inst_mat[1]

for zwd, zwd_inst_list in zip([1, 4], zwd_inst_mat):
    prop_matched_list = []
    for zwd_inst in zwd_inst_list:
        # print(i)
        if zwd_inst.desc_list is None:
            # print("{} of zwd({}) has no features".format(zwd_inst.ix, zwd_inst.zwd))
            prop_matched_list.append(0)
            continue
        len_zwd_inst_features = len(zwd_inst.desc_list)
        all_matches = flann.knnMatch(inst.desc_list, zwd_inst.desc_list, k=len_zwd_inst_features)
        matches_mat = np.array([
            [m.distance for m in sorted(
                [match for match in row], key=lambda x: x.trainIdx
            )] for row in all_matches]
        )
        prop_matched_list.append(
            np.sum(np.where(matches_mat < DISTANCE_TH, 1, 0).sum(axis=0) != 0) / len_zwd_inst_features
        )
    
    # write match result to data fame
    prop_matched_list.sort(reverse=True)
    df_sift_train_out.ix[inst.ix, "z{}_top1".format(zwd):"z{}_top{}".format(zwd, TOP_K)] = prop_matched_list[:TOP_K]
    df_sift_train_out.ix[inst.ix, "z{}_mean".format(zwd)] = np.mean(prop_matched_list)
````

### Speed up

````python
## TODO: 改良 knnMatch 不要全部會更快
def zwd_inst_match(zwd_inst):
    if zwd_inst.desc_list is None:
        # print("{} of zwd({}) has no features".format(zwd_inst.ix, zwd_inst.zwd))
        return 0
    len_zwd_inst_features = len(zwd_inst.desc_list)
    all_matches = flann.knnMatch(inst.desc_list, zwd_inst.desc_list, k=len_zwd_inst_features)
    matches_mat = np.array([
        [m.distance for m in sorted(
            [match for match in row], key=lambda x: x.trainIdx
        )] for row in all_matches]
    )
    return (
        np.sum(np.where(matches_mat < DISTANCE_TH, 1, 0).sum(axis=0) != 0) / len_zwd_inst_features
    )
````

````python
def zwd_whole_combine(inst):
    for zwd, zwd_inst_list in zip([1, 4], zwd_inst_mat):
        prop_matched_list = map(zwd_inst_match, zwd_inst_list)
        prop_matched_list.sort(reverse=True)
        # black magic, count index position
        top_start_ix = (zwd - 1) * TOP_K + 1
        top_end_ix = zwd * TOP_K
        mean_ix = 12 * TOP_K + zwd
        df_sift_train_out.ix[inst.ix, top_start_ix:top_end_ix] = prop_matched_list[:TOP_K]
        df_sift_train_out.ix[inst.ix, mean_ix] = np.mean(prop_matched_list)
````
