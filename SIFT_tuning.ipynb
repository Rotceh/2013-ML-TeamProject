```python
import pandas as pd
import cv2
from matplotlib import pylab as pl
from matplotlib import cm
import numpy as np
```

```python
%matplotlib inline
```

```python
from util.get_data import get_new_suite
```

```python
df_train, df_test, train_Y, train_X, test_Y, test_X = get_new_suite(suite_name="60x60grey")
```

```python
group_train = df_train.groupby("y")
```

```python
def rand_choice_ix_bygroup(grouped, n=1):
    """return a list of random index nparray from each group
    
    Parameters
    ==========
    grouped 
        a pandas DataFrameGroupBy object
    n: 1, int
        number of indexes of each group
    """
    return [np.random.choice(
        grouped.get_group(gk).index,
        size=n,
        replace=False
    ) for gk in grouped.groups.keys()]
```

```python
rand_ix_array = rand_choice_ix_bygroup(group_train, n=8)
```

```python
PIXEL_X = 60
PIXEL_Y = 60

def grey_convert(wd_array):
    """convert [0, 1] greyscale to [0, 255] (dark, white)"""
    return np.round((1 - wd_array) * 255).astype(np.uint8)
```

```python
sigma = 1.4 # default: 1.6
contrastThreshold = 0.03 # default: 0.04
nOctaveLayers = 6  # default: 3
edgeThreshold = 5 # default: 10

sift = cv2.SIFT(
    sigma=sigma, 
    contrastThreshold=contrastThreshold,
    nOctaveLayers=nOctaveLayers,
    edgeThreshold=edgeThreshold
)
```

```python
%%time
mixed_bucket = [
    sift.detectAndCompute(
        grey_convert(x_arr.reshape(PIXEL_X, PIXEL_Y)),
        None
    ) for x_arr in train_X[np.ravel(rand_ix_array)]
]
kp_bucket = [mixed_item[0] for mixed_item in mixed_bucket]
desc_bucket = [mixed_item[1] for mixed_item in mixed_bucket]
```

```python
n_subfig_x = 12
n_subfig_y = 8
px_scale = 2
n_subfig = n_subfig_x * n_subfig_y
f, ax_mat = pl.subplots(
    n_subfig_y, n_subfig_x, 
    figsize=(n_subfig_x * px_scale, n_subfig_y * px_scale)
)
ax_mat = ax_mat.T.flatten()


for i, (ix, ax) in enumerate(zip(np.ravel(rand_ix_array), ax_mat)):
    img_sp = cv2.drawKeypoints(
        grey_convert(train_X[ix]).reshape(PIXEL_X, PIXEL_Y), 
        kp_bucket[i],
        flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
    )
    ax.imshow(img_sp, aspect=1)
    ax.set_adjustable('box-forced')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("i: {}".format(ix))
```

```python
f.savefig(
    "results/figs/train_sift_allzoidac_60x60_.png", 
    dpi=150,
    transparent=True,
    bbox_inches='tight',
    pad_inches=0
)
```
