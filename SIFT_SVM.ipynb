```python
from __future__ import print_function
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.cross_validation import cross_val_score, train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
```

```python
CSV_TRAIN = "dataset/train_60x60_grey_SIFT_top100_distth300.csv"
df_train = pd.read_csv(CSV_TRAIN)
```

```python
Y = df_train.y.values
X = df_train.iloc[:, 1:].values
```

### Grid CV

```python
GRID_RESULT = "results/grid_CV/svm_60x60greysift_0109_2.txt"

param_list = [
    {'C': [0.1, 0.2, 0.5, 1, 2, 5, 10, 100, 500, 1000]},
    {'kernel': ['rbf'],
     'C': [100, 200, 300, 400, 500],
     'gamma': [0.1, 0.05, 0.02,
               0.01, 0.005, 0.002,
               0.001]}
]

linsvm_clf = svm.LinearSVC()
svm_clf = svm.SVC(cache_size=2000)  # Don't try this at home!

train_X, test_X, train_Y, test_Y = train_test_split(
    X, Y, test_size=0.25, random_state=5566
)
```

```python
%%capture out_cap
for clf, param in zip((linsvm_clf, svm_clf), param_list):
    grid_cv = GridSearchCV(clf, param, n_jobs=15, verbose=1)
    grid_cv.fit(train_X, train_Y)
    
    with open(GRID_RESULT, "a") as f:
        msg = (
            "\n\n>>>>> [best parameter from grid searching] <<<<<\n"
            "%s\n\n\n" % grid_cv.best_estimator_
        )
        print(msg)
        print(msg, file=f)
        
        for params, mean_score, scores in grid_cv.grid_scores_:
            row_msg = (
                "%0.3f (+/-%0.03f) for %r"
                % (mean_score, scores.std() / 2, params)
            )
            print(row_msg)
            print(msg, file=f)

    # best model to predict
    pred_Y = grid_cv.predict(test_X)
    
    # prediction report
    print(classification_report(test_Y, pred_Y))
    print(confusion_matrix(test_Y, pred_Y))
```
