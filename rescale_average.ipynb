## Gaussian Threshold

```python
def grey_convert(wd_array):
    """convert [0, 1] greyscale to [0, 255] (dark, white)"""
    inv_arr = 1 - wd_array
    return np.round(np.where(inv_arr < 0, 0, inv_arr) * 255).astype(np.uint8)
```

```python
len(group_px_mean)
```

```python
import cv2
PX_X = 60
PX_Y = 60
```

```python
px_mean_adaptive_th = []

for px_mean in group_px_mean:
    img = grey_convert(px_mean).reshape(60, 60)
    px_mean_adaptive_th.append(
        cv2.adaptiveThreshold(img, 255,
                              cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                              cv2.THRESH_BINARY, 15, 0.2)
    )
```

```python
n_subfig_x = 6 # 12 生肖
n_subfig_y = 2
px_scale = 2
n_subfig = n_subfig_x * n_subfig_y

f, ax_mat = pl.subplots(
    n_subfig_y, n_subfig_x, 
    figsize=(n_subfig_x * px_scale, n_subfig_y * px_scale)
)
ax_mat = ax_mat.flatten()
for ix, (ax, px_avg) in enumerate(zip(ax_mat, px_mean_adaptive_th), 1):
    ax.imshow(px_avg.reshape(60, 60), cm.binary_r)
    ax.set_adjustable('box-forced')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("class: {}".format(ix))
pl.show()
```

### Save only mean

```python
def df_combine_y_front(x_only, y, num_x):
    df_out = pd.DataFrame(
        x_only,
        columns=[
            "x_{}".format(i+1) for i in range(num_x)
        ]
    )
    df_out['y'] = y
    
    # rearrange columns
    df_out = df_out[['y'] + ["x_{}".format(i+1) for i in range(num_x)]]
    
    return df_out
```

```python
df_mean_only = df_combine_y_front(group_px_mean, [int(k) for k in group_train.groups.keys()], 60 * 60)
```

```python
CSV_MEAN_ONLY = "dataset/train_60x60_bw_mean_only.csv"
df_mean_only.to_csv(CSV_MEAN_ONLY, index=False)
```
