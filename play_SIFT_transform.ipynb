````python
def rand_choice_ix_bygroup(grouped, n=1):
    """return a list of random index nparray from each group
    
    Parameters
    ==========
    grouped 
        a pandas DataFrameGroupBy object
    n: 1, int
        number of indexes of each group
    """
    return [np.random.choice(
        grouped.get_group(gk).index,
        size=n,
        replace=False
    ) for gk in group_train.groups.keys()]
````

````python
# combined multiple nparray into one
combined_ix_array = np.c_[part_ix_array].flatten()
````

````python
PIXEL_X = 60
PIXEL_Y = 60

def grey_convert(wd_array):
    """convert [0, 1] greyscale to [0, 255] (dark, white)"""
    return np.round((1 - wd_array) * 255).astype(np.uint8)
````

````python
sigma = 1.6 # default: 1.6
# If your image is captured with a weak camera with soft lenses,
# you might want to reduce the number.
contrastThreshold = 0.04 # default: 0.04
nOctaveLayers = 3  # default: 3
edgeThreshold = 5 # default: 10
sift = cv2.SIFT(
    sigma=sigma, 
    contrastThreshold=contrastThreshold,
    nOctaveLayers=nOctaveLayers,
    edgeThreshold=edgeThreshold
)
````

````python
%%time
mixed_bucket = [
    sift.detectAndCompute(
        grey_convert(x_arr.reshape(PIXEL_X, PIXEL_Y)),
        None
    ) for x_arr in train_X[combined_ix_array]
]
kp_bucket = [mixed_item[0] for mixed_item in mixed_bucket]
desc_bucket = [mixed_item[1] for mixed_item in mixed_bucket]
````

````python
n_subfig_x = 9
n_subfig_y = 3
px_scale = 2
n_subfig = n_subfig_x * n_subfig_y
f, ax_mat = pl.subplots(
    n_subfig_y, n_subfig_x, 
    figsize=(n_subfig_x * px_scale, n_subfig_y * px_scale)
)
ax_mat = ax_mat.T.flatten()

        
for i, (ix, ax) in enumerate(zip(combined_ix_array, ax_mat)):
    img_sp = cv2.drawKeypoints(
        grey_convert(train_X[ix]).reshape(PIXEL_X, PIXEL_Y), 
        kp_bucket[i],
        flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
    )
    ax.imshow(img_sp, aspect=1)
    ax.set_adjustable('box-forced')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("i: {}".format(ix))
````

### Matching

````python
# flatten desc
flatten_desc = np.c_[[desc for desc_list in desc_bucket for desc in desc_list]]
````

````python
i = -6 # take ith data
ix = combined_ix_array[i]
cur_ix_desc_list = desc_bucket[i]
````

````python
# FLANN parameters
K_SCALE = 10
SIMILAR_TOP_K = 50 * K_SCALE # 跟該生肖相同的數量 x K_SCALE
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=100)   # or pass empty dictionary

flann = cv2.FlannBasedMatcher(index_params, search_params)
````

````python
%%time
cur_all_matches = flann.knnMatch(cur_ix_desc_list, flatten_desc, k=SIMILAR_TOP_K)
````
