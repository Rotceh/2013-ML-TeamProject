````python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
````

````python
import pandas as pd
import numpy as np
````

````python
CSV_ZERO_TRAIN = "dataset/train_na2zero.csv"
````

````python
df = pd.read_csv(CSV_ZERO_TRAIN, nrows=1000)
````

目前看起來一組資料的長寬為：

- W: 122
- H: 105

生肖按照「鼠、牛、虎、兔、龍、蛇、馬、羊、猴、雞、狗、豬」排列。

````python
PIXEL_X = 122
PIXEL_Y = 105
zodiac_char = list("鼠牛虎兔龍蛇馬羊猴雞狗豬")
zodiac_map = {
    i : zodiac_char[i-1] 
    for i in range(1, 13)
}
````

````python
i_th = 5
row = df.ix[i_th, :]
y = int(row[0])
print("生肖：{}({})".format(zodiac_map[y], y))
px_array = row[1:].values.reshape(PIXEL_X, PIXEL_Y)
````

````python
plt.figure()
plt.subplot(121)
plt.imshow(px_array, aspect="equal", cmap=cm.binary)

plt.subplot(122)
plt.imshow(px_array, aspect="equal", cmap=cm.binary_r)
plt.show()
````

````python
N_SUBFIG_X = 8
N_SUBFIG_Y = 2
f, ax_array = plt.subplots(N_SUBFIG_Y, N_SUBFIG_X, figsize=(16, 4))

ax_array.shape = N_SUBFIG_X * N_SUBFIG_Y
for ax in ax_array:
    ax.imshow(px_array, aspect=1, cmap=cm.binary)
    ax.set_adjustable('box-forced')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("hi")

f.subplots_adjust()
# f.savefig("hihi.png", dpi=300, transparent=True)
````

````python
group = df.groupby(by="y")
````

````python
group.size()
````

````python
n_subfig_x = 8
n_subfig_y = 2
n_subfig = n_subfig_x * n_subfig_y

f, ax_array = plt.subplots(N_SUBFIG_Y, N_SUBFIG_X, figsize=(n_subfig_x*1.5, n_subfig_y*1.5))
ax_array.shape = n_subfig

y = 12  # 生肖
g = group.get_group(y)
print("生肖：{}({})".format(zodiac_map[int(y)], int(y)))
ix_array = np.random.choice(g.index, size=n_subfig)

for ix, ax in zip(ix_array, ax_array):
    row = df.ix[ix, :]
    px_array = row[1:].values.reshape(PIXEL_X, PIXEL_Y)
    ax.imshow(px_array, aspect=1, cmap=cm.binary)
    ax.set_adjustable('box-forced')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.set_title("i: {}".format(ix))

f.subplots_adjust()
# f.savefig("hihi.png", dpi=300, transparent=True)
````
