````python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
````

````python
import pandas as pd
import numpy as np
````

````python
CSV_ZERO_TRAIN = "dataset/train_na2zero.csv"
````

````python
df = pd.read_csv(CSV_ZERO_TRAIN, nrows=1000)
````

目前看起來一組資料的長寬為：

- W: 122
- H: 105

生肖按照「鼠、牛、虎、兔、龍、蛇、馬、羊、猴、雞、狗、豬」排列。

````python
PIXEL_X = 122
PIXEL_Y = 105
zodiac_char = list("鼠牛虎兔龍蛇馬羊猴雞狗豬")
zodiac_map = {
    i : zodiac_char[i-1] 
    for i in range(1, 13)
}
````

````python
i_th = 5
row = df.ix[i_th, :]
y = int(row[0])
print("生肖：{}({})".format(zodiac_map[y], y))
px_array = row[1:].values.reshape(PIXEL_X, PIXEL_Y)
````

````python
plt.figure()
plt.subplot(121)
plt.imshow(px_array, aspect="equal", cmap=cm.binary)

plt.subplot(122)
plt.imshow(px_array, aspect="equal", cmap=cm.binary_r)
plt.show()
````

````python
group = df.groupby(by="y")
````

````python
group.size()
````

````python
def plot_zodiac_sample(zodiac_ix=1, n_subfig_x=8, n_subfig_y=3, px_scale=1.5):
    n_subfig = n_subfig_x * n_subfig_y
    f, ax_array = plt.subplots(
        n_subfig_y, n_subfig_x, 
        figsize=(n_subfig_x * px_scale, n_subfig_y * px_scale)
    )
    ax_array.shape = n_subfig
    
    g = group.get_group(zodiac_ix)
    print("生肖：{}({})".format(zodiac_map[zodiac_ix], zodiac_ix))
    ix_array = np.random.choice(g.index, size=n_subfig)
    
    for ix, ax in zip(ix_array, ax_array):
        row = df.ix[ix, :]
        px_array = row[1:].values.reshape(PIXEL_X, PIXEL_Y)
        ax.imshow(px_array, aspect=1, cmap=cm.binary)
        ax.set_adjustable('box-forced')
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.set_title("i: {}".format(ix))
    
    # f.subplots_adjust()
    return f
````

````python
%%time
ix_list = list(range(1, 6))  # 可以把它改成 13 就會做全部的生肖
zodiac_fig_list = [plot_zodiac_sample(zodiac_ix, n_subfig_y=1) for zodiac_ix in ix_list]
````

可以透過 `f.savefig("mypng.png", dpi=300, transparent=True, bbox_inches='tight', pad_inches=0)` 來儲存圖檔。  
底下的指令用來產生 repo 內的檔案（建議更改檔名跑跑看，因為是亂數所以結果不會相同）。

蓋到的話可以使用 `git checkout HEAD -- results/figs/train_pixelview*.png`（這會讓本機所有更動消除）  
細節見 <http://stackoverflow.com/questions/7147270> 。

````python
for ix, fig in zip(ix_list, zodiac_fig_list):
    fig.savefig(
        "results/figs/train_pixelview_{}({})_8x2.png".format(
            zodiac_map[ix], ix
        ), dpi=150, transparent=True,
        bbox_inches='tight',  # 去除討厭的外框
        pad_inches=0
    )
````
