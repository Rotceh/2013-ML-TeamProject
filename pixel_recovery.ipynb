````python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
````

````python
import pandas as pd
import numpy as np
````

````python
CSV_ZERO_TRAIN = "dataset/train_na2zero.csv"
````

````python
df = pd.read_csv(CSV_ZERO_TRAIN, nrows=100)
````

````python
plt.imshow(px_array, cmap=cm.binary)
````
