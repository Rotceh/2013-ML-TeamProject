```python
import numpy as np
import pandas as pd
import cv2
```

```python
from util.resize import resize_60x60
```

```python
CSV_TRAIN = "dataset/train_na2zero.csv"
CSV_TEST = "dataset/test_na2zero.csv"
df_train = pd.read_csv(CSV_TRAIN)
df_test = pd.read_csv(CSV_TEST)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
PIXEL_X = 122
PIXEL_Y = 105
zodiac_char = list("鼠牛虎兔龍蛇馬羊猴雞狗豬")
zodiac_map = {
    i : zodiac_char[i-1] 
    for i in range(1, 13)
}
```

```python
group_train = df_train.groupby(by="y")
```

```python
group_train.size()
```

```python
def df_row_iter(df):
    for i, row in df.iterrows():
        yield int(row.iat[0]), row.iloc[1:].values
```

```python
%%time
df_out = pd.DataFrame(
    [resize_60x60(px_array).flatten() for _, px_array in df_row_iter(df_train)],
    columns=[
        "x_{}".format(i+1) for i in range(60 * 60)
    ]
)
```

```python
df_out['y'] = df_train.y
```

```python
# rearrange columns
df_out = df_out[['y'] + ["x_{}".format(i+1) for i in range(60 * 60)]]
```

```python
df_out.ix[:3, 1230:1240]
```

```python
n_subfig_x = 12 # 12 生肖
n_subfig_y = 6
px_scale = 1.5
n_subfig = n_subfig_x * n_subfig_y
f, ax_mat = plt.subplots(
    n_subfig_y, n_subfig_x, 
    figsize=(n_subfig_x * px_scale, n_subfig_y * px_scale)
)
ax_mat = ax_mat.T

# precompute the index array so we can reuse after resizing
rand_ix_array = [
    np.random.choice(
        group_train.get_group(i).index,
        size=n_subfig_y,
        replace=False
    ) for i in range(1, 13)
]

for ix_array, (ix, ax_array) in zip(rand_ix_array, enumerate(ax_mat, 1)):
    g = group_train.get_group(ix)
    # print("生肖：{}({})".format(zodiac_map[ix], ix))
    # ix_array = np.random.choice(g.index, size=n_subfig_y, replace=False)
    for ix, ax in zip(ix_array, ax_array):
        row = df_train.ix[ix, :]
        px_array = row[1:].values.reshape(PIXEL_X, PIXEL_Y)
        ax.imshow(px_array, aspect=1, cmap=cm.binary)
        ax.set_adjustable('box-forced')
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.set_title("i: {}".format(ix))
```

```python
def df_transform_60x60(df):
    df_out = pd.DataFrame(
        [resize_60x60(px_array).flatten() for _, px_array in df_row_iter(df)],
        columns=[
            "x_{}".format(i+1) for i in range(60 * 60)
        ]
    )
    df_out['y'] = df.y
    
    # rearrange columns
    df_out = df_out[['y'] + ["x_{}".format(i+1) for i in range(60 * 60)]]
    
    return df_out
```

```python
df_train_out = df_transform_60x60(df_train)
df_test_out = df_transform_60x60(df_test)
```

```python
CSV_OUT_TRAIN = "dataset/train_zero_60x60.csv"
CSV_OUT_TEST = "dataset/test_zero_60x60.csv"

df_train_out.to_csv(CSV_OUT_TRAIN, index=False)
df_test_out.to_csv(CSV_OUT_TEST, index=False)
```
